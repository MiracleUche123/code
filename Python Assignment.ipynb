```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randn (1,50,30)
for a in x:
  y = (a * 2)
  w = (a ** 3)
    
plt.figure(figsize = (12,6))

plt.plot(x, 'r*', label = 'a')
plt.plot(y, 'g-', label = 'y')
plt.plot(w, 'b--', label = 'w')
plt.title('Matplotlib Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend();
```
